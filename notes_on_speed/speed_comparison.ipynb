```python
import pandas as pd
import matplotlib.pyplot as plt

# File paths
file_paths = [
    'spock_wave_fd4_gaussian',
    'spock_wave_matmul_double_gaussian',
    'spock_wave_matmul_single_gaussian',
    'spock_wave_fd4_light_halo',
    'spock_wave_matmul_single_pol_light_halo',
    'spock_wave_matmul_single_order_14_light_halo',
    'spock_hybrid_fd4_order_8_light_halo',
    'spock_hybrid_fd4_order_8_single_light_halo',
    'spock_hybrid_matmul_double_pol_light_halo',
    'spock_hybrid_matmul_double_order_8_light_halo',
    'spock_hybrid_matmul_double_order_14_light_halo',
    'spock_hybrid_matmul_single_pol_light_halo',
    'spock_hybrid_matmul_single_order_8_light_halo',
    'spock_hybrid_matmul_single_order_8_single_light_halo',
    'spock_hybrid_matmul_single_order_14_light_halo',
]

# Column names based on the provided format
columns = [
    'Time', 'Step', 'dt', 'NCell', 'NUpdate_Cell', 'ElapsedTime',
    'Perf_Overall', 'Perf_PerRank', 'NUpdate_Lv0', 'NUpdate_Lv1',
    'NUpdate_Lv2', 'NUpdate_Lv3', 'NUpdate_Lv4', 'NUpdate_Lv5',
    'NUpdate_Lv6', 'NUpdate_Lv7', 'NUpdate_Lv8', 'NUpdate_Lv9'
]

# Initialize a dictionary to store average Perf_PerRank for each file
averages = {}

for file_path in file_paths:
    # Read the data
    df = pd.read_csv(file_path, delim_whitespace=True, header=None, skiprows=1, names=columns).iloc[-30:]

    # Calculate the average Perf_PerRank
    avg_perf_per_rank = df['Perf_PerRank'].mean()

    # Store the average in the dictionary
    averages[file_path] = avg_perf_per_rank

# Convert the dictionary to a DataFrame for plotting
avg_df = pd.DataFrame(list(averages.items()), columns=['File', 'Average_Perf_PerRank'])

# Plot the results
plt.figure(figsize=(10, 6))
plt.barh(avg_df['File'], avg_df['Average_Perf_PerRank'], color='skyblue')
plt.xlabel('Average Perf_PerRank')
plt.title('Average Perf_PerRank for Different Files')
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
